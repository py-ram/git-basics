```python
letters = ["a", "b", "c", "d", "e", "f", "g"]

# Last 3 items
last_three = letters[-3:]
print(last_three)  # ["e", "f", "g"]
```

```python

# Everything except last 2
except_last = letters[:-2]
print(except_last)  # ["a", "b", "c", "d", "e"]
```

```python
# From -5 to -2 (not including -2)
middle = letters[-5:-2]
print(middle)  # ["c", "d", "e"]
```

```python
car = {
    "brand": "Toyota",
    "model": "Camry",
    "year": 2020,
    "color": "blue"
}

# pop() - Remove and return value
model = car.pop("model")
print(model)  # "Camry"
print(car)    # {"brand": "Toyota", "year": 2020, "color": "blue"}
```

```python
# pop() with default (doesn't crash if key missing)
owner = car.pop("owner", "Unknown")
print(owner)  # "Unknown"
```

```python

# del - Delete item
del car["color"]
print(car)  # {"brand": "Toyota", "year": 2020}

# clear() - Remove all items
car.clear()
print(car)  # {}
```

```python
students = {
    "roll 1": {
        "Name": "Alice",
        "age": 20,
        "grade": 85,
        "courses": ["Math", "Physics"]
    },
    "Bob": {
        "age": 21,
        "grade": 92,
        "courses": ["Chemistry", "Biology"]
    }
}


count_dict = 1
for key, value in students.items():
    if isinstance(value, dict):
        for key1, value1 in value.items():
            count_dict += 1

print(count_dict)
```

```python
values = (1, 2, 3, 2, 4, 2, 5)

# count() - Count occurrences
twos = values.count(2)
print(twos)  # 3

# index() - Find position of first occurrence
position = values.index(3)
print(position)  # 2

# index() with start position
position2 = values.index(2, 2)  # Find 2 starting from index 2
print(position2)  # 3
```
