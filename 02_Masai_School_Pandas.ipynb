```python
# Find rows where age is missing
missing_age = titanic[titanic["age"].isna()]
print(f"Passengers with unknown age: {len(missing_age)}")   # 177

# Find rows where age is NOT missing
known_age = titanic[titanic["age"].notna()]
print(f"Passengers with known age: {len(known_age)}")       # 714

# Check if this group has a different survival rate
print(known_age["survived"].mean().round(2))    # 0.41
print(missing_age["survived"].mean().round(2))  # 0.29
# Interesting  passengers with unknown ages survived less often
```

```python
"Find first-class women who survived,
 sort them by what they paid,
 give me a clean table with just sex, age, and fare."
```

```python
# Step 1  Filter: first class, female, survived
result = titanic[
    (titanic["pclass"] == 1) &
    (titanic["sex"] == "female") &
    (titanic["survived"] == 1)
]

# Step 2  Select only the columns we need
result = result[["sex", "age", "fare"]]

# Step 3  Sort by fare, highest first
result = result.sort_values("fare", ascending=False)

print(result.head(10))
print(f"\nTotal: {len(result)} first-class female survivors")
```

```python
# The clean one-liner version
result = (
    titanic[
        (titanic["pclass"] == 1) &
        (titanic["sex"] == "female") &
        (titanic["survived"] == 1)
    ]
    [["sex", "age", "fare"]]
    .sort_values("fare", ascending=False)
)

print(result.head(10))
```
