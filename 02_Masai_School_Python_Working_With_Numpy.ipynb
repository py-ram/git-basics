```python
# Using regular Python lists
list1 = [1, 2, 3, 4, 5]
list2 = [10, 20, 30, 40, 50]

# What happens if we try to add them?
result = list1 + list2
print(result)
# Output: [1, 2, 3, 4, 5, 10, 20, 30, 40, 50]
# This CONCATENATES the lists, not element-wise addition!

# To do element-wise addition with lists, we need a loop:
result = []
for i in range(len(list1)):
    result.append(list1[i] + list2[i])
print(result)
# Output: [11, 22, 33, 44, 55]
```

```python
arr = np.array([1, 2, 3, 4, 5])

# Addition
print(arr + 10)
# Output: [11 12 13 14 15]

# Subtraction
print(arr - 2)
# Output: [-1  0  1  2  3]

# Multiplication
print(arr * 3)
# Output: [ 3  6  9 12 15]

# Division
print(arr / 2)
# Output: [0.5 1.  1.5 2.  2.5]

# Power
print(arr ** 2)
# Output: [ 1  4  9 16 25]

# Modulus
print(arr % 2)
# Output: [1 0 1 0 1]
```

```python
mat1 = np.array([[1, 2], [3, 4]])
mat2 = np.array([[5, 6], [7, 8]])

# Element-wise operations
print(mat1 + mat2)
# Output:
# [[ 6  8]
#  [10 12]]
```

```python
print(mat1 * mat2)  # Element-wise multiplication
# Output:
# [[ 5 12]
#  [21 32]]
```

```python
# Matrix multiplication (dot product)
print(np.dot(mat1, mat2))
# Output:
# [[19 22]
#  [43 50]]
```

```python
arr = np.array([1, 2, 3, 4, 5])

print(np.sum(arr))        # 15
print(arr.sum())          # 15 (method syntax)
print(np.mean(arr))       # 3.0
print(np.std(arr))        # 1.4142135623730951 (standard deviation)
print(np.var(arr))        # 2.0 (variance)
print(np.min(arr))        # 1
print(np.max(arr))        # 5
print(np.argmin(arr))     # 0 (index of minimum)
print(np.argmax(arr))     # 4 (index of maximum)
```

```python
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])

print(arr_2d.sum())              # 21 (sum of all elements)
print(arr_2d.sum(axis=0))        # [ 5  7  9] (sum along rows - column sums)
print(arr_2d.sum(axis=1))        # [ 6 15] (sum along columns - row sums)

print(arr_2d.mean(axis=0))       # [2.5 3.5 4.5]
print(arr_2d.max(axis=1))        # [3 6]
```

```python
import numpy as np
import time

# Create large data
size = 1000000

python_list = list(range(size))
numpy_array = np.arange(size)

# Test 1: Summing elements
# Python list
start = time.time()
total = sum(python_list)
print(total)
python_time = time.time() - start

# NumPy array
start = time.time()
total = np.sum(numpy_array)
print(total)
numpy_time = time.time() - start

print(f"Python list sum: {python_time:.5f} seconds")
print(f"NumPy array sum: {numpy_time:.5f} seconds")
print(f"NumPy is {python_time/numpy_time:.1f}x faster")
```
