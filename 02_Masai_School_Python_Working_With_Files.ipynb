# Looping throug the files

```python
file1 = open("/content/fruits.txt", "r")

for line in file1:
    print(line)

file1.close()
```

```python
with open("/content/fruits.txt", "r") as balaji:
    content = balaji.read()
    print(content)
```

```python
# No need to close the file
# automatically close the file
# safer, cleaner, and industry staandard.
```

# Writing to a file.

```python
with open("/content/fruits.txt", "w") as balaji:
    balaji.write("Hello there, this is a new line")
    balaji.write("I am super hero")
```

```python
with open("/content/fruits.txt", "a") as balaji:
    balaji.write('''\n We are Netherdlands
    \n New Line in Gemrany''')
```

```python
with open("dfgdfv/fruits.txt", "a") as balaji:

    name = input("Enter your name: ")
    balaji.write(f"\n  Your Name is {name}")
```

```python
with open("/content/fruits.txt", "r") as balaji:

    name = input("Enter your name: ")
    balaji.write(f"\n  Your Name is {name}")
```

# read + write

```python
file = open("fruits.txt", "r+")

content = file.read()
print("Before Writing:")
print(content)

file.write("\nNew Line Added")

file.close()
```

```python
file = open("fruits.txt", "r+")

file.seek(10)   # Move cursor to beginning
file.write("Jayanth")

file.close()
```

```python
with open("fruits.txt", "r+") as file:
    file.seek(10)   # Move cursor to beginning
    content = file.write("venkatesh")
    print(content)
    file.close()
```
